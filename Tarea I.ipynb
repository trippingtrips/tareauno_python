## Version Control (GIT)

Carácteristicas de los VCS: 
 +  Son herramientas que podemos usar para rastrear cambios a los origenes del código u otras colecciones de archivos y folders.
 + Mantiene un historial de cambios
 + Facilita la colaboración con otros usuarios
 + Mantienen metadata de quien creo cada snapshot, mensajes asociados a este snapshot y más.



¿Por qué son importantes los VCS?

 + Nos permite ver nuestras versiones antiguas del proyecto
 + Trabajar paralelamente en diferentes versiones de nuestro proyecyo
 + El valor que le da a nuestro trabajo ver los cambios que han hecho nuestros compañeros y como han resuelto conflictos

**Los modelos de data de Git:** Según la lectura, hay varios tipos de enfoque que podemos tomar para el control de versiones, el historial y la colaboración.

*Snapshots*: Git tiene una colección de archivos y folders como una serie de snapshots. En terminología de git un archivo se le llama *Blob* , a un directorio se le llama *tree* y a los snapshots que vienen de ramas paralelas *commits*.

En Git, los snapshots practicamente se derivan de del conjunto de padres que lo precedieron, escribimos padres ya que un snapshot puede venir de más de un origen. Cuando se está referenciando otros objetos (trees, blobs or commits) no se referencia directamente a ellos sino que se utiliza algo llamado *hash*, este es un código que cada uno tiene. Los SHA-1 hasges son dificiles de memorizar así que git nos puede apoyar teniendo referencias.
